```
#Ques12 - Which team won the toss most of the time?
df['toss_winner'].value_counts().idxmax()
```

```
#Ques14 - Which team won most matches by runs?
n = df[df['won_by'] == 'Runs']
n.groupby('match_winner').size().idxmax()
```
